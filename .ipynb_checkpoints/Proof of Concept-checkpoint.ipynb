### Data Gathering and Preview

```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# folder path
folder_path = "NBA Shot Data"

# load and sort all csv files
csv_files = sorted(glob.glob(os.path.join(folder_path, "*.csv")))

# read and combine into single df
df = pd.concat(
    [pd.read_csv(file) for file in csv_files],
    ignore_index=True
)
```

```python
# preview
print("Total rows:", len(df))
print(df.head())
```

### Proof of Concept - Time Series Model (AR(1))

```python
import pandas as pd
import numpy as np

# copy df to avoid SettingWithCopyWarning
df_copy = df.copy()

# most recent 5 seasons
recent_seasons = ['2019-20','2020-21','2021-22','2022-23','2023-24']

# filter dataset for the last 5 seasons
df_recent = df_copy[df_copy['SEASON_2'].isin(recent_seasons)]
print("Total rows after filter:", len(df_recent))
print("Unique players in last 5 seasons:", df_recent['PLAYER_NAME'].nunique())
```

```python
# convert SHOT_MADE to binary
df_recent['SHOT_MADE_BINARY'] = df_recent['SHOT_MADE'].astype(int)

# get unique players
players = df_recent['PLAYER_NAME'].unique()

# store results
results = []

for player in players:
    # filter shots for this player
    player_df = df_recent[df_recent['PLAYER_NAME'] == player].sort_values('GAME_DATE')
    
    # skip players with very few shots (at least 20)
    if len(player_df) < 20:
        continue
    
    # create lagged series for AR(1)
    player_df['SHOT_LAG1'] = player_df['SHOT_MADE_BINARY'].shift(1)
    player_ar = player_df.dropna(subset=['SHOT_LAG1'])
    
    Y = player_ar['SHOT_MADE_BINARY'].values
    X = player_ar['SHOT_LAG1'].values
    
    # add intercept
    X_matrix = np.column_stack((np.ones(len(X)), X))
    
    # estimate closed form solution
    beta = np.linalg.inv(X_matrix.T @ X_matrix) @ (X_matrix.T @ Y)
    
    # store results
    results.append({
        'PLAYER_NAME': player,
        'TOTAL_SHOTS': len(player_df),
        'AR1_COEFFICIENT': beta[1]
    })

# convert results to df
ar_results_df = pd.DataFrame(results)

# preview top 10 players by AR(1) coefficient
print(ar_results_df.sort_values('AR1_COEFFICIENT', ascending=False).head(10))
```

```python
# select top 5 players by AR(1) coefficient
top_players = ar_results_df.sort_values('AR1_COEFFICIENT', ascending=False).head(5)['PLAYER_NAME'].values

for player_name in top_players:
    player_df = df_recent[df_recent['PLAYER_NAME'] == player_name].sort_values('GAME_DATE')
    
    # binary shot series
    shot_series = player_df['SHOT_MADE_BINARY']
    
    # skip if too few shots
    if len(shot_series) < 4:
        continue
    
    # determine max lags
    max_lags = min(20, len(shot_series)//2)
    
    plt.figure(figsize=(12,5))
    
    # ACF plot
    plt.subplot(1,2,1)
    plot_acf(shot_series, lags=max_lags, ax=plt.gca())
    plt.title(f'ACF for {player_name}')
    
    # PACF plot
    plt.subplot(1,2,2)
    plot_pacf(shot_series, lags=max_lags, ax=plt.gca(), method='ywm')
    plt.title(f'PACF for {player_name}')
    
    plt.tight_layout()
    plt.show()
```
