```python
import pandas as pd
import glob
import os

# folder path
folder_path = "NBA Shot Data"

# load and sort all csv files
csv_files = sorted(glob.glob(os.path.join(folder_path, "*.csv")))

# read and combine into single df
df = pd.concat(
    [pd.read_csv(file) for file in csv_files],
    ignore_index=True
)

# preview
print("Total rows:", len(df))
print(df.head())
```
